```python
from md4.train import _get_checkpoint_manager
from md4.configs.hollow_md4 import text8
import orbax.checkpoint
import jax

# Get config and set model type
config = text8.get_config()
config.model_type = "hollow"  # Make sure we're using the hollow model type

# Create checkpoint manager
checkpoint_manager = _get_checkpoint_manager(
    config, "/root/md4/expt_contantlr_deeper__model_untie"
)

# Create a simple dictionary structure matching the expected checkpoint structure
params_dict = {"CondEmbedding_0": {}, "Embed_0": {}, "HollowTransformer_0": {}}

# Create restore args
restore_args = orbax.checkpoint.RestoreArgs(
    dtype=jax.numpy.dtype("float32")  # or whatever dtype your model uses
)

# Try to restore just the parameters
try:
    restored = checkpoint_manager.restore(
        300000, items={"params": params_dict}, restore_kwargs={"params": restore_args}
    )
    params = restored["params"]
    print("Successfully loaded parameters")
    print(jax.tree_map(lambda x: x.shape if hasattr(x, "shape") else None, params))
except Exception as e:
    print(f"Error loading checkpoint: {e}")
```
